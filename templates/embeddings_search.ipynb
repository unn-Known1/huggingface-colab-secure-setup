## Step 2: Load Embedding Model

```python
from sentence_transformers import SentenceTransformer
import numpy as np

print(f"Loading {MODEL_ID}...")

embedder = SentenceTransformer(MODEL_ID)

print(f"\u2705 Model loaded!")
print(f"   Embedding dimension: {embedder.get_sentence_embedding_dimension()}")
```

## Step 3: Create Embeddings

```python
# Your documents - CHANGE THIS!
documents = [
    "Python is a high-level programming language known for its simplicity.",
    "Machine learning is a subset of artificial intelligence that enables systems to learn.",
    "Deep learning uses neural networks with multiple layers to solve complex problems.",
    "Natural language processing helps computers understand human language.",
    "Computer vision enables computers to interpret and understand visual information.",
    "Transformers are the architecture behind modern language models like BERT and GPT.",
    "GitHub is a platform for version control and collaboration.",
    "Docker helps developers containerize their applications."
]

print(f"Creating embeddings for {len(documents)} documents...\n")

# Generate embeddings
embeddings = embedder.encode(documents, show_progress_bar=True)
embeddings = np.array(embeddings)

print(f"\u2705 Embeddings created!")
print(f"   Shape: {embeddings.shape}")
```

## Step 4: Semantic Search

```python
from sklearn.metrics.pairwise import cosine_similarity

def semantic_search(query, top_k=3):
    """
    Find most similar documents to a query.
    
    Args:
        query: Search query
        top_k: Number of results to return
    """
    # Embed query
    query_embedding = embedder.encode([query])
    
    # Calculate similarities
    similarities = cosine_similarity(query_embedding, embeddings)[0]
    
    # Get top results
    top_indices = similarities.argsort()[-top_k:][::-1]
    
    results = []
    for idx in top_indices:
        results.append({
            "document": documents[idx],
            "similarity": similarities[idx]
        })
    
    return results

print("\u2705 Search function ready!")
```

## Step 5: Try It!

```python
# Search query - CHANGE THIS!
query = "How do neural networks work?"

print(f"\ud83d\udd0d Query: {query}\n")
print("-" * 60)

results = semantic_search(query, top_k=3)

for i, result in enumerate(results, 1):
    print(f"\n{i}. Score: {result['similarity']:.3f}")
    print(f"   {result['document']}")
```

## Step 6: Build a Simple RAG System
