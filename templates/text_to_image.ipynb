## ⚙️ Configuration

```python
# Model selection
MODEL_ID = "stabilityai/stable-diffusion-xl-base-1.0"  # @param ["stabilityai/stable-diffusion-xl-base-1.0", "runwayml/stable-diffusion-v1-5", "stabilityai/stable-diffusion-2-1"]

# Generation settings
NUM_INFERENCE_STEPS = 25  # @param {type:"slider", min:10, max:50, step:5}
GUIDANCE_SCALE = 7.5  # @param {type:"slider", min:1, max:15, step:0.5}
IMAGE_SIZE = 1024  # @param [512, 768, 1024]

print(f"\ud83d\udc49 Model: {MODEL_ID}")
print(f"\ud83d\udc49 Steps: {NUM_INFERENCE_STEPS}")
print(f"\ud83d\udc49 Guidance: {GUIDANCE_SCALE}")
```

## Step 3: Load Model

```python
import torch
from diffusers import StableDiffusionXLPipeline

print(f"\ud83d\ude80 Loading {MODEL_ID}...")

# Load pipeline
pipe = StableDiffusionXLPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    use_safetensors=True
)

# Move to GPU
pipe = pipe.to("cuda")
pipe.enable_attention_slicing()  # Save memory

print("\u2705\u2705 Model loaded! Ready to generate!\u2705\u2705")
```

## Step 4: Generate Images

```python
def generate_image(prompt, negative_prompt="", num_images=1):
    """
    Generate images from text prompt.
    
    Args:
        prompt: Text description of desired image
        negative_prompt: What you DON'T want in the image
        num_images: Number of images to generate
    """
    images = pipe(
        prompt=prompt,
        negative_prompt=negative_prompt,
        num_inference_steps=NUM_INFERENCE_STEPS,
        guidance_scale=GUIDANCE_SCALE,
        height=IMAGE_SIZE,
        width=IMAGE_SIZE,
        num_images_per_prompt=num_images
    ).images
    
    return images

print("\u2705 Generation function ready!")
print("\ud83d\udc47 Use: generate_image('your prompt')")
```

## Step 5: Try It Out!

```python
# Your prompt here - CHANGE THIS!
PROMPT = "A serene mountain lake at sunrise, crystal clear water, snow-capped peaks, beautiful reflections, photorealistic, 8k ultra detailed"

NEGATIVE_PROMPT = "blurry, low quality, distorted, ugly, watermark"

print(f"\ud83d\udce2 Generating: {PROMPT[:50]}...")
print("\u23f3 This may take 1-2 minutes...\n")

# Generate
images = generate_image(PROMPT, NEGATIVE_PROMPT)

# Display
from IPython.display import display
for i, img in enumerate(images):
    display(img)
```
