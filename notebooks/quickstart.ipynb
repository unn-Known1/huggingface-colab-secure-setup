## Step 1: Configure Secure Environment

Before loading any models, we configure the environment for maximum privacy and security.

```python
# Configure environment variables for privacy and security
import os

# Disable telemetry - we don't want any data sent to Hugging Face
os.environ["HF_HUB_DISABLE_TELEMETRY"] = "1"

# Set secure cache directory
os.environ["HF_HOME"] = "/content/.hf_cache"
os.environ["TRANSFORMERS_CACHE"] = "/content/.hf_cache"

# Disable tokenizer parallelism to prevent race conditions
os.environ["TOKENIZERS_PARALLELISM"] = "false"

# Create cache directory
!mkdir -p /content/.hf_cache

print("\u2705 Secure environment configured!")
```

## Step 3: Load Your First Model

Now let's load a model. We'll use a small, fast model for demonstration.

```python
from transformers import AutoModel, AutoTokenizer

# Use GPT-2 as a simple example - no authentication required
model_name = "gpt2"

print(f"Loading model: {model_name}...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

print(f"\u2705 Model loaded successfully!")
print(f"   Model type: {type(model).__name__}")
print(f"   Parameters: {model.num_parameters():,}")
```

## Step 4: Run Inference

Now let's use the model for text generation.

```python
from transformers import AutoModelForCausalLM

# Load for text generation
gen_model = AutoModelForCausalLM.from_pretrained("gpt2")

# Your input text
input_text = "The future of AI security is"

# Tokenize
inputs = tokenizer(input_text, return_tensors="pt")

# Generate
outputs = gen_model.generate(
    **inputs,
    max_new_tokens=50,
    do_sample=True,
    temperature=0.9
)

# Decode and display
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"Input: {input_text}")
print(f"\nGenerated: {generated_text}")
```

## Step 5: Memory Cleanup (Security Best Practice)

Always clear sensitive data from memory after use.

```python
import gc

# Clear memory after processing
def secure_cleanup():
    """Securely clear memory containing sensitive data."""
    gc.collect()
    try:
        import torch
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
            torch.cuda.synchronize()
    except:
        pass
    print("\u2705 Memory cleared securely!")

# Call cleanup
secure_cleanup()
```
